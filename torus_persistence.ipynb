```python
import matplotlib.pyplot as plt
from gudhi.datasets.generators import points

# Generate 50 points randomly on a torus in R^6
gen_points = points.torus(n_samples = 50, dim = 2)

# gen_points = points.sphere(n_samples = 50, ambient_dim = 2, radius = 1, sample = "random")

print(gen_points)

# Unpack coordinates
x = [p[0] for p in gen_points]
y = [p[1] for p in gen_points]

# Plot the points
plt.scatter(x, y, color='purple', s=30)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('50 Points on a Torus (1D Circle in R²)')
plt.gca().set_aspect('equal')  # Keep aspect ratio square
plt.grid(True)
plt.show()
```

```python
import gudhi as gd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from statsbombpy import sb
import re
import os
import warnings
import numpy as np
from itertools import combinations
import random
from gudhi.hera import wasserstein_distance
import pandas as pd
import seaborn as sns

rips_complex = gd.RipsComplex(points=gen_points, max_edge_length=1000.0)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=3)

diag = simplex_tree.persistence()

persistence_intervals = simplex_tree.persistence_intervals_in_dimension(0)
gd.plot_persistence_diagram(diag)
#gd.plot_persistence_barcode(diag)
plt.title("2-Torus Persistence Diagram")
plt.savefig('2torus_persistence_diagram.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
